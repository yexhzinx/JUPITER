```python
# ----------------------------------------------------------------------
# 이상치 탐지 (Anomaly Detection) 개념 정리
# ----------------------------------------------------------------------

# 1. 기본 개념 및 목적
# - 정의: 데이터셋 내에서 대다수의 관측치와 현저하게 다르거나 일치하지 않는 패턴(이상치, Outlier)을 식별하는 프로세스.
# - 이상치의 의미: 시스템 오작동, 사기 행위, 오류, 또는 흥미로운 새로운 현상을 나타낼 수 있음.
# - 목적: 희귀 이벤트를 경고하거나, 분석에서 이상치를 제외하여 모델 성능을 개선.

# 2. 이상치의 유형 (Type of Outliers)
# - 전역 이상치 (Global Outlier): 데이터셋 전체와 비교할 때 비정상적인 값.
# - 국소 이상치 (Local Outlier): 주변 이웃 데이터 포인트와 비교할 때 비정상적인 값.
# - 문맥적 이상치 (Contextual Outlier): 특정 상황(시간, 위치 등)의 맥락에서 비정상적인 값.

# ----------------------------------------------------------------------
# 주요 활용 알고리즘 및 특징 (개념 위주)
# ----------------------------------------------------------------------

# A. 분리 (Isolation) 기반
# - Isolation Forest: 무작위 트리를 반복 생성하여 이상치를 빠르게 고립(isolate)시킴.
#   - 특징: 고속, 대용량 데이터에 적합, 이상치가 정상 데이터보다 트리의 루트에 가까움.

# B. 밀도 및 거리 기반
# - LOF (Local Outlier Factor): 데이터 포인트가 주변 이웃에 비해 얼마나 희소한지(국소 밀도) 측정.
#   - 특징: 국소적 이상치 탐지에 효과적, 밀도가 균일하지 않은 데이터셋에 유용.
# - DBSCAN: 군집에 속하지 않고 밀집 영역 밖에 있는 데이터를 노이즈(이상치)로 식별.
#   - 특징: 밀도 기반으로, 불규칙한 모양의 군집에서도 이상치를 식별 가능.

# C. 단일 클래스 학습 기반
# - One-Class SVM: 정상 데이터(Normal Data)만 학습하여 정상 패턴의 경계를 정의.
#   - 특징: 경계 밖에 있는 새로운 데이터를 이상치로 판단, 정상 패턴이 명확할 때 주로 사용.

# D. 통계 및 확률 기반
# - GMM (Gaussian Mixture Model): 데이터 분포를 여러 개의 정규분포 혼합으로 추정.
#   - 특징: 모델이 추정한 분포의 확률이 매우 낮은 영역의 데이터를 이상치로 판단, 확률적 해석 가능.
```
