```python
# 1️⃣ 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# 2️⃣ 데이터 불러오기 및 기본 확인
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print("Train 데이터 크기:", train.shape)
print("Test 데이터 크기:", test.shape)

print("\n[Train 데이터 정보]")
train.info()
print("\n[Test 데이터 정보]")
test.info()
```

```python
# 3️⃣ 결측치 확인
print("\n[Train 결측치 개수]")
print(train.isnull().sum())
print("\n[Test 결측치 개수]")
print(test.isnull().sum())# 4️⃣ 결측치 처리
# - Item_Weight : 최소값으로 채움
# - Outlet_Size : 최빈값으로 채움
train['Item_Weight'] = train['Item_Weight'].fillna(train['Item_Weight'].min())
train['Outlet_Size'] = train['Outlet_Size'].fillna(train['Outlet_Size'].mode()[0])

test['Item_Weight'] = test['Item_Weight'].fillna(test['Item_Weight'].min())
test['Outlet_Size'] = test['Outlet_Size'].fillna(test['Outlet_Size'].mode()[0])
```

```python
# 5️⃣ 필요 없는 컬럼 제거
# Item_Identifier는 단순 식별자이므로 제거
train = train.drop('Item_Identifier', axis=1)
test = test.drop('Item_Identifier', axis=1)
```

```python

# 6️⃣ 범주형(문자형) 변수 인코딩
# - 학습용과 테스트용을 합쳐 LabelEncoder 적용 → 동일 기준 유지
cols = train.columns[train.dtypes == 'object']
for col in cols:
    le = LabelEncoder()
    le.fit(pd.concat([train[col], test[col]]))
    train[col] = le.transform(train[col])
    test[col] = le.transform(test[col])
```

```python
# 7️⃣ 독립변수(X), 종속변수(y) 분리
X = train.drop('Item_Outlet_Sales', axis=1)
y = train['Item_Outlet_Sales']
```

```python
# 8️⃣ 데이터 분할 (학습용 / 검증용)
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
# 9️⃣ 선형회귀 모델 학습
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 🔟 예측
y_pred = lr.predict(X_val)
```

```python
# 11️⃣ 평가 (MSE, MAE, RMSE, R2)
mse = mean_squared_error(y_val, y_pred)
mae = mean_absolute_error(y_val, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred)

print("\n[모델 성능 평가]")
print(f"MSE : {mse:.2f}")
print(f"MAE : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R2 : {r2:.4f}")
```

```python
# 12️⃣ 시각화 (실제값 vs 예측값)
plt.figure(figsize=(6,4))
sns.scatterplot(x=y_val, y=y_pred)
plt.xlabel("실제값 (y_val)")
plt.ylabel("예측값 (y_pred)")
plt.title("실제값 vs 예측값 (Linear Regression)")
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

train = pd.read_csv('train.csv')

# 결측치 제거 또는 채우기
train['Item_Weight'] = train['Item_Weight'].fillna(train['Item_Weight'].min())
train['Outlet_Size'] = train['Outlet_Size'].fillna(train['Outlet_Size'].mode()[0])

# 독립변수(X), 종속변수(y)
X = pd.get_dummies(train['Item_Type'], drop_first=True)  # 범주형 변수 더미화
y = train['Item_Outlet_Sales']

# 상수항 추가
X = sm.add_constant(X)

# OLS 회귀분석
model = sm.OLS(y, X).fit()

# 결과 출력
print(model.summary())
```
